```python
def binary_search(searchlist,key):
    
    length = len(searchlist)
    low = 0
    upper = length - 1
    
    while (low<=upper):
        middle = (low + upper)//2
        
        if (key == searchlist[middle]):
            return middle
        
        elif (key < searchlist[middle]):
            upper = middle - 1
            
        elif (key > searchlist[middle]):
            low = middle+1
    else:
        return False
    
key = 85
a = [-1,2,4,38,45,64,85]

print(binary_search(a,key))

```
